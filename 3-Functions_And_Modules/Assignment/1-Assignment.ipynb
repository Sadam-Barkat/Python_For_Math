```python
import math
```

```python
def distance(x1,y1,x2,y2):
  d = (x2 - x1)**2 + (y2 - y1)**2
  d = math.sqrt(d)
  return d
```

```python
x1 = 3
y1 = 4
x2 = 7
y2 = 1
print(distance(x1,y1,x2,y2))
```

```python
def hypotenues(a , b):
  h = a**2 + b**2
  h = math.sqrt(h)
  return h
```

```python
print(hypotenues(6,8))
```

```python
def permutataion(n, k):
  p = math.perm(n,k)
  return p
```

```python
print(permutataion(5,3))
```
