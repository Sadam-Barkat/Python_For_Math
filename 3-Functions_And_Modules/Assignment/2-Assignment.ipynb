```python
import math
```

```python
def arithmetic_sum(a, d):
  n = 10
  Sum = n/2 *((2*a + (n - 1) )*d)
  return Sum
```

```python
a = 5
d = 3
print(arithmetic_sum(a,d))
```

```python
def combination(n, r):
  c = math.comb(n, r)
  return c
```

```python
n = 7
r = 3
print(combination(n, r))
```

```python
def geometric_sum(a, r, n):
    if r == 1:
        return a * n
    else:
        return a * (1 - r**n) / (1 - r)
```
