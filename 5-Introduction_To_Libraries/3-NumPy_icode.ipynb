##### Linear Algebra

The np.matmul(a, b) function performs matrix multiplication between a and b.
For matrix multiplication, the number of columns in the first matrix a must equal the number of rows in the second matrix b.

```python
import numpy as np

# 2D Matrix Multiplication
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print(A)
print(B)
```

```python
result = np.matmul(A, B)
print("Matrix Multiplication with matmul:\n", result)
```

```python
import numpy as np

# Define 3D arrays
A = np.array([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])

B = np.array([
    [[9, 10], [11, 12]],
    [[13, 14], [15, 16]],[[3,4],[5,6]]])


print("Array A:\n", A)
print("Array B:\n", B)
# Perform matrix multiplication
result = np.matmul(A, B)

print("Resulting 3D Array:\n", result)
```

```python
#Determinant of a Matrix
C = np.array([[1,2,5],[3,4,8],[8,7,6]])

det = np.linalg.det(C)
print(int(det))
```

```python
# Find the determinant
c = np.identity(5)
np.linalg.det(c)
```

```python
c = np.full((5,5),[8,7,6,9,4])
print(c)
print(np.linalg.det(c))
```

```python
# Example: Inverse of a Matrix
D = np.array([[1, 2], [3, 4]])
inv = np.linalg.inv(D)
print("Inverse of the Matrix:\n", inv)
```

```python
D = np.array([[1, 2], [3, 4]])
E = np.array([5, 6])
print(np.linalg.solve(D,E))
```

```python
np.matmul(np.linalg.inv(D),E)
```
