#Question No 3

**Normal Division**

in normal division compiler divide two numbers and give the output in float no

sign use (/)

example

```python
a = 5
b = 2
print(a/b)
```

**Floor Division**

in floor division compiler divide the two numbers and then roundup the output

like if output is 3.5 then it will roundup the output in 4 and if the output is 3.4 then it will roundup it into 3 because 4 is less then by 5

and the sign is use (//)

exmaple

```python
a = 7
b = 3
print(a//b)
```
