# **Basic Calculation Exercises**

```python
# 1. Addition

num1 = 5
num2 = 10
sum_result = num1 + num2
print(f"Addition: num1 + num2 = ",sum_result)
```

```python
# 2. Subtraction

num3 = 20
num4 = 7
sub_result = num3 - num4
print(f"Subtraction: num3 - num4 = ",sub_result)
```

```python
# 3. Multiplication

num5 = 4
num6 = 3
mul_result = num5 * num6
print(f"Multiplication: num5 * num6 = ",mul_result)
```

```python
# 4. Division

num7 = 16
num8 = 2
div_result = num7 / num8
print("Division: num7 / num8 = ",div_result)
```

```python
# 5. Modulus (Remainder)

num9 = 15
num10 = 4
mod_result = num9 % num10
print("Modulus: num9 % num10 = ",mod_result)
```

```python
# 6. Exponentiation

num11 = 2
num12 = 3
exp_result = num11 ** num12
print("Exponentiation: num11 ^ num12 = ",exp_result)
```

```python
# 7. Floor Division

num13 = 17
num14 = 3
floor_div_result = num13 // num14
print("Floor Division: num13 // num14 = ",floor_div_result)
```

```python
import cmath  # To handle complex numbers

# User Input
print("Quadratic Equation Solver")
print("For the equation of the form ax^2 + bx + c = 0")

# Input coefficients
a = float(input("Enter coefficient a: "))
b = float(input("Enter coefficient b: "))
c = float(input("Enter coefficient c: "))

# Calculate the discriminant using pow()
discriminant = pow(b, 2) - 4 * a * c

# Calculate the roots using the quadratic formula
root1 = (-b + cmath.sqrt(discriminant)) / (2 * a)
root2 = (-b - cmath.sqrt(discriminant)) / (2 * a)

# Output the roots
print("Root 1:", root1)
print("Root 2:", root2)
```

```python
# Formatting Numbers
number = float(input("Enter a floating-point number: "))
formatted_number = "{:.2f}".format(number)
print("Formatted number: {}".format(formatted_number))
```

```python
# f-string: Creating a Simple Invoice
item_name = input("Enter the item name: ")
quantity = int(input("Enter the quantity: "))
price_per_item = float(input("Enter the price per item: "))
total = quantity * price_per_item
print(f"Invoice:\nItem: {item_name}\nQuantity: {quantity}\nTotal: ${total:.2f}")
```
