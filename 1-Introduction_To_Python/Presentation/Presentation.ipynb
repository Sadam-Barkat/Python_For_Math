#What is a Variable?
A variable is a container for storing data values.
In Python, variables are created when you assign a value to them.

#Key Characteristics:
Variables do not need to be declared.
Variables are case-sensitive.

```python
x = 10     # integer variable
name = "Ali"  # string variable
pi = 3.14  # float variable
X = 19
print(x, X)
```

#Rules for Naming Variables:
Must start with a letter (a-z, A-Z) or underscore (_).

Cannot start with a number.

Can contain alphanumeric characters (a-z, A-Z, 0-9) and underscores.

Case-sensitive (name and Name are different).

#Examples of Valid and Invalid Variable Names:
Valid: name, _age, price1

Invalid: 1name, @price, total-cost

#Single Assignment:
Directly assign a value to a variable.

#Multiple Assignment:
Assign values to multiple variables in a single line.

```python
# Single assignment
x = 5
y = "Hello"

# Multiple assignment
a, b, c = 1, 2, 3
print(a,b,c)
```

#Types of Variables:
Integer (e.g., x = 5)

Float (e.g., pi = 3.14)

String (e.g., name = "Alice")

Boolean (e.g., is_student = True)

List, Tuple, Dictionary, Set (collections)

```python
is_active = True        # Boolean
price = 19.99           # Float
fruits = ["apple", "banana"] # List
print(type(price))
```

#Checking Variable Type:
Use type() to get the type of a variable.

#Type Casting:
Converting a variable from one type to another (e.g., int to float).

```python
x = 10
print(type(x))  # Output: <class 'int'>

# Type casting
x = str(x)      # Convert integer to string
y = float(x)    # Convert string to float
```

#Dynamic Typing:
Variables can change type after assignment.
#Example:
Start with an integer, then reassign as a string.

```python
x = 5           # x is an integer
print(type(x))  # Output: <class 'int'>

x = "Hello"     # Now x is a string
print(type(x))  # Output: <class 'str'>
print(type(x))
print(type(x))
```

#What are Constants?
Constants are variables that should not change after assignment.

#Python Convention:
Use uppercase letters to indicate constants.

```python
PI = 3.14159
GRAVITY = 9.8
```

#Local Variables:
Defined inside a function and only accessible within that function.
#Global Variables:
Defined outside any function and accessible throughout the code.

```python
# Global variable
x = "global"

def my_func():
    # Local variable
    x = "local"
    print("Inside function:", x)

my_func()
print("Outside function:", x)
```
