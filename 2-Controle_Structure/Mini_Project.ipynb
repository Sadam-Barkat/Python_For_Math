#Guess Number Mini Project Using Python

```python
import random

secret = random.randint(0,99)
guess = int(input("Guess the number : "))
count = 0

while guess != secret:
  if guess > secret:
    print("Number is too high")
  else:
    print("Number is too low")
  guess = int(input("Guess the number : "))
  count += 1

print(f'Congratulations you found the number {guess} and you try {count} times')

```
