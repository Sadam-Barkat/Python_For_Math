**Q2**: Write a Python program that:

1-Asks the user for their working hours (a whole number).

2-Asks the user to enter their ratings for three tasks (each rating is out of 100).


The rules for getting a bonus are:

1-The worker needs to have worked at least 40 hours.

2-The average rating for the three tasks must be at least 60.

If both conditions are met, print "Bonus Granted". Otherwise, print "Bonus Denied".

```python
hours = int(input("Enter the working hours : "))
sum_ratings = 0
i = 0
while i < 3:
  ratings = int(input("Enter the ratings for three week :"))
  if ratings > 100:
    print("rating is out of range")
  sum_ratings = ratings + sum_ratings
  i += 1
avg_rating = sum_ratings / 3

if avg_rating >= 60 and hours >=40:
  print("Bonus Granted")
else:
  print("Bonus Denied")
```

**Q4**: Write a Python program that asks the user for two integers. If both numbers are even, multiply them, otherwise, add them. Use nested if statements and logical operators to check the number conditions

```python
num1 = int(input("Enter first number : "))
num2 = int(input("Enter second number : "))

if num1 % 2 == 0:
  if num2 % 2 == 0:
    mul = num1 * num2
    print("Both are even numbers : ",mul)
  else:
    add = num1 + num2
    print("One number is even : ",add)
else:
  add = num1 + num2
  print("Both numbers are odd : ",add)
```

**Q5**: Write a program that asks the user for the number of terms in the Fibonacci sequence they want to generate. Use a for loop to compute and print the sequence up to that number of terms.

```python
num = int(input("Enter the number for fibonachi sequence : "))
z = 0
x = 0
y = 1

for i in range(num):
  print(z,end=" , ")
  z = x + y
  x = y
  y = z
```

**Q6**: Below is a Python program intended to print all prime numbers between 2 and 30. However, the code contains errors and does not function as expected. Identify the errors, explain what they are, and provide the corrected code.

```python
# Python program to print prime numbers between 2 and 30

n = 30

for num in range(2, n + 1):
    count = 0
    for i in range(1, n):
        if num % i == 0:
            count += 1
    if count == 2:
      print(num,end=" ")
```
