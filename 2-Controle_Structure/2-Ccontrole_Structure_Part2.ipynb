# **Loops (For and While)**

## **For Loop**

```python
#Counting Steps on a Staircase
for step in range(1, 11):  # 1 to 10
    print(f"Step {step}")
```

```python
sentence = "Learning Python is fun!"
vowel_count = 0 #1

for char in sentence:
    if char.lower() in "aeiou":
        vowel_count += 1

print("Number of vowels:", vowel_count)
```

```python
sentence = "Learning Python is fun!"
vowel_count = 0

for i in range(len(sentence)):
    if sentence[i].lower() in "aeiou":
        vowel_count += 1

print("Number of vowels:", vowel_count)
```

```python
#Checking All Items in a Grocery List
grocery_list = ["Apples", "Bananas", "Milk", "Eggs"]
for item in grocery_list:
    print(f"I need to buy {item}")
```

## **While Loop**

```python
#Saving Money for a Goal

savings = 0
goal = 100
week = 1

while savings < goal:
    savings += 10  # saving $10 each week
    print(f"Week {week}: Saved ${savings}")
    week += 1
```

```python
#Battery Level Check

battery = 20

while battery < 100:
    battery += 10  # charging by 10% each time
    print(f"Charging... Battery at {battery}%")
```

# **Control Flow Statements**

## **Break**

```python
for num in range(10):
    if num == 6:
        print("Found 6! Byee Byee!")
        break
    print(num)
```

## **Continue**

```python
for num in range(20):
    if num % 2 == 0:
        continue
    print(num)
```

## **Pass**

```python
for num in range(20):
    if num % 2 == 0:
      pass
```

# **if and if-elif ladder**

```python
temperature = 18

if temperature > 25:
    print("It's hot outside!")
if temperature > 15:
    print("It's warm outside!")
if temperature > 5:
    print("It's cool outside!")
else:
    print("It's cold outside!")
```

```python
temperature = 18

if temperature > 25:
    print("It's hot outside!")
elif temperature > 15:
    print("It's warm outside!")
elif temperature > 5:
    print("It's cool outside!")
else:
    print("It's cold outside!")
```
