#Presentation On Conditional Statements

**1.Conditional Statements**

**Definition**: Conditional statements are used to execute code based on specific conditions.

```python
# if statement
x = 10
if x > 5:
    print("x is greater than 5")

# if-else statement
x = 3
if x > 5:
    print("x is greater than 5")
else:
    print("x is less than or equal to 5")

# if-elif-else statement
x = 15
if x < 10:
    print("x is less than 10")
elif x == 10:
    print("x is equal to 10")
else:
    print("x is greater than 10")
```

**2.Logical Operators**

**Definition**: Combine multiple conditions.
and: True if both conditions are true.
or: True if at least one condition is true.
not: True if the condition is false.

```python
# Using 'and'
x = 7
if x > 5 and x < 10:
    print("x is between 5 and 10")

# Using 'or'
x = 3
if x < 5 or x > 15:
    print("x is outside the range of 5 to 15")

# Using 'not'
is_raining = False
if not is_raining:
    print("You can go outside without an umbrella")
```

**3. Mini-Project: Simple Banking System**

Project: Create a program that simulates a simple banking system where a user can check balance, deposit, or withdraw money with conditions.

```python
balance = 1000  # Initial balance

while True:
    print("1. Check Balance")
    print("2. Deposit")
    print("3. Withdraw")
    print("4. Exit")
    choice = int(input("Choose an option: "))

    if choice == 1:
        print(f"Your current balance is ${balance}")
    elif choice == 2:
        amount = float(input("Enter deposit amount: "))
        if amount > 0:
            balance += amount
            print(f"${amount} deposited. New balance is ${balance}")
        else:
            print("Invalid amount.")
    elif choice == 3:
        amount = float(input("Enter withdrawal amount: "))
        if amount > 0 and amount <= balance:
            balance -= amount
            print(f"${amount} withdrawn. New balance is ${balance}")
        else:
            print("Invalid amount or insufficient funds.")
    elif choice == 4:
        print("Thank you for using the banking system.")
        break
    else:
        print("Invalid choice.")
```
